```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import vonmises, norm, cauchy

from modules.params import *
from modules.potential import *
from modules.aux_functions import *
from modules.free_energy_calc import *
from modules.plot_functions import *
from modules.simulation_framework import model_simulation
```

```python
sigma_list = [round(0.33 + 0.02 * i, 2) for i in range(26)]
t_end = 50
print(sigma_list)

tp_list = []
vmax_list = []
vmax_t_list = []

for sigma_l in sigma_list: 
    u_store_fp, vmax_time, vmax, vmax_idx, u_max, u_end = model_simulation(sigma=sigma_l, t_end=t_end, c_a=c_a, l_a=l_a, c_r=c_r, l_r=l_r)
    free_energy, entropy, interaction = calc_free_energy(u_store_fp, sigma=sigma_l)
    min_dt_time = plot_free_energy_with_tp_clustering(u_store_fp, vmax_time, sigma=sigma_l, t_end=t_end, c_a=c_a, l_a=l_a, c_r=c_r, l_r=l_r)
    
    tp_list.append(min_dt_time)
    vmax_list.append(vmax)
    vmax_t_list.append(vmax_time)
```

```python
fig, ax1 = plt.subplots()
ax1.scatter(sigma_list, tp_list, marker='x', color = 'b', label='data points')
ax1.plot(sigma_list, tp_list, color='tab:orange', linestyle=':', label='tie line')
ax1.set_xlabel(r'$\sigma$')
ax1.set_ylabel(r'$t$')
ax1.legend()
ax1.grid()
ax1.set_ylim(bottom=0)
ax1.set_title(r'Time of reaching turning point in $F[\rho]$ for different $\sigma$' + '\n' + 'in clustering regime')
plt.savefig(plot_dir + 'tp_time_clustering.pdf')
plt.show()


fig, ax2 = plt.subplots()
ax2.scatter(sigma_list, vmax_list, marker='x', color = 'b', label='data points')
ax2.plot(sigma_list, vmax_list, color='tab:orange', linestyle=':', label='tie line')
ax2.set_xlabel(r'$\sigma$')
ax2.set_ylabel(r'sup $\rho(x,t)$')
ax2.legend()
ax2.grid()
ax2.set_title(r'sup $\rho(x,t_{end})$ for different $\sigma$ in clustering regime' + '\n' + r'$t_{end}$ = ' + str(t_end))
plt.savefig(plot_dir + 'max_dens_values_clustering.pdf')
plt.show()
```
